## normalization  using Z score: 

using Z-Score reduces the impact of extreme values because it standardizes data based on the mean and standard deviation. Unlike Min-Max, which is sensitive to outliers (as it compresses data to [0, 1] based on the min and max values), Z-Score centers data around 0 and scales it by variability, ensuring fair contribution from all data points, even in the presence of outliers. This makes it ideal for RFM analysis and clustering methods like K-Means.

Elbow Method is used to find the best 𝐾
K value by plotting SSE (Sum of Squared Errors).
SSE (Inertia) measures how compact the clusters are—the lower the SSE, the better the clustering.

```python
from sklearn.preprocessing import StandardScaler
rfm=df2_user[["R","F","M"]]
scaler= StandardScaler()
rfm_scaled=scaler.fit_transform (rfm)
rfm_scaled_df=pd.DataFrame(rfm_scaled,columns=["R","F","M"],index= df2_user.index)
df3_scaled=df2_user[['CustomerID']].join(rfm_scaled_df)
print(df3_scaled.head())
```

```python
inertia=[]
k_values=range(1,11)
for k in k_values:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.plot(k_values,inertia, marker="o")
plt.title("elbow for best K")
plt.xlabel("number of Cluster(K)")
plt.ylabel("interia(sse)")
plt.show()

```

this best k is for the whole data, but we can sheck for each R,F,M, to check if 4 is best k

```python
def show_elbow(df):
    inertia=[]
    k = range(1,11)
    for i in k:
        kmeans= KMeans(n_clusters=i)
        kmeans.fit(df)
        inertia.append(kmeans.inertia_)

    plt.plot(k,inertia, marker="o")
    plt.title("elbow method")
    plt.xlabel("number of cluster")
    plt.ylabel("inertia")
    plt.show()
```

K-Means requires a 2D array because it calculates distances between data points in a feature space. Each row represents a sample, and each column represents a feature. If the input is 1D, K-Means cannot determine whether it's multiple samples with one feature or one sample with multiple features. using double  brackets[['R']] make sure it is 2D

```python
# use [["r"]] make it to be 2D,
show_elbow(df3_scaled[['R']])
```

```python
def order_cluster(df,cluster_col,target_col,ascending= False):
    cluster_means=df.groupby(cluster_col)[target_col].mean().reset_index()
    cluster_means=cluster_means.sort_values(by=target_col).reset_index(drop=True)
    # using the new index and cluster_col order to build an dictionary
    cluster_mapping= {row[cluster_col]: idx for idx,row in cluster_means.iterrows()}
    df[cluster_col]=df[cluster_col].map(cluster_mapping)
    return df
```

- If you want to highlight the characteristics of each category in the RFM dimensions → Radar Chart
- If you want to visually see the distribution of customers in the RFM dimensions → 3D Scatter Plot
- If you want to analyze the proportion of each category → Stacked Bar Chart

```python
def plot_custom_radar_chart(df4):
    # Group data by 'Totalvalue' and compute the mean of R_valued, F_valued, M_valued
    df_mean = df4.groupby('Totalvalue')[['R-valued', 'F_valued', 'M_valued']].mean()
    features = ['R-valued', 'F_valued', 'M_valued']
    
    # Number of features and categories
    num_features = len(features)
    angles = np.linspace(0, 2 * np.pi, num_features, endpoint=False).tolist()
    angles += angles[:1]  # Close the radar chart loop
    
    # Initialize radar chart
    fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))
    
    # Plot data for each customer group
    for category in df_mean.index:
        values = df_mean.loc[category].tolist()
        values += values[:1]  # Close the loop
        ax.plot(angles, values, label=category, linewidth=2)
        ax.fill(angles, values, alpha=0.25)

    # Add feature labels
    ax.set_thetagrids(np.degrees(angles[:-1]), features)
    ax.set_title("Radar Chart: Customer Segmentation", fontsize=16)
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1), title="Customer Groups")
    plt.show()

# Call the radar chart function
plot_custom_radar_chart(df4)
```

This radar chart illustrates the RFM characteristics of four customer groups: High Value, Lost Customer, Low Value, and Medium Value. The High Value group has high frequency and monetary values with recent transactions, making them the most valuable customers. The Lost Customer group shows low frequency and monetary values with transactions made a long time ago, indicating they need re-engagement. The Low Value group has moderate recency but low frequency and spending, requiring minimal investment. The Medium Value group has moderate engagement across all dimensions, with potential for growth through targeted strategies.
